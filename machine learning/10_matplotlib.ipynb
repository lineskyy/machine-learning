# matplotlib
- 파이썬의 대표적인 과학 계산용 그래프 라이브러리
- plt.figure: 축과 그래프, 그래프의 제목 같은 것들을 표시할 수 있는 컨테이너
- plt.axes: 눈금과 레이블, 그래프가 있는 테두리 박스로 시각화를 형성하는 좌표평면

```python
plt.figure() # 그래프를 그리기 위한 틀을 만든다.
plt.axes() # 그 안에 그래프들을 그린다.
plt.show() # 실제 화면에 표시한다.
```

## 단일 그래프

- plot 함수
    - line plot을 그려주는 함수
    - figure, axes 객체가 없다면 이를 만들어서 그려준다.

```python
plt.plot(np.arange(2,7)) # 1개만 들어갈 경우 y값 (x값은 인덱스 번호)
plt.show()
```

## 다중 그래프
- 1개의 axes 안에 여러개의 그래프를 그린다.

```python
plt.figure(figsize=[15,5]) # [넓이, 높이]
plt.plot(np.arange(5)) # x 축
plt.plot(np.arange(2,7))    # y 축
plt.show()
```

## subplots 함수를 이용하여 axes 여러개 그리기

```python
fig, ax = plt.subplots(2,2,figsize =[15,5])
print(ax.shape)
ax[0,0].plot(np.arange(2,6))
ax[0,0].plot(np.arange(5))

ax[1,0].plot(np.arange(5))
ax[1,0].plot(np.arange(5))
```

- 다른방식으로 여러개 그리기

```python
arr = np.arange(5)
plt.figure(figsize=[15,5])
plt.subplot(2,2,1)
plt.plot(np.arange(5))
plt.subplot(2,2,2)
plt.plot(arr)

plt.subplot(2,2,3)
plt.plot(arr)

plt.subplot(2,2,4)
plt.plot(arr)

plt.show()
```

# 타이틀 설정하기

```python
fig,ax = plt.subplots(1,2,figsize=(15,5))
fig.suptitle("figure title")

x = range(0,10) # x축
x.np.exp()
ax[0].plot(x.np.exp(x) # y축
ax[0].set_title("exp")

x = range(1,1000) # x축
y = np.log(x) # y축
ax[1].plot(x,y)
ax[1].set_title("log")
plt.show
```

- x 와 y축에 이름넣기

```python
x = range(1,100)
y = np.log(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel("x축이름")
ax.set_ylabel("y축이름")
plt.show()
```

## 눈금회전하기

```python
fig,ax = plt.subplots()
ax.plot(x,y)
ax.tick_params(axis="x",labelrotation=90)
plt.show()
```

## x와 y 축 눈금 지정하기
- `set_xticks , set_yticks`

```python
fig, ax = plt.subplots(figsize=[10,10])
ax.plot(range(50))
ax.set_xticks(range(50))
ax.set_yticks(range(50))
ax.tick_params(axis="x",labelrotation=90)
plt.show()
```

## 범례 표시하기
- axes 객체의 legend 메소드

# 데이터 불러오기

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

```python
df.columns =["지불금액","팁금액","성별","흡연자여부","요일","식사타임","인원수"]
df.info()
```

# Line Plot
-일반적으로 시간을 x축으로 두고y축 데이터에 추이를 볼때 사용

```python
tmp = df.groupby("요일").agg({"지불금액":"sum"}).reset_index()
```

- 요일별 매출에 대한 추이 확인하기

```python
fig,ax = plt.subplots()
ax.plot(tmp["요일"],tmp["지불금액"])
plt.show()
```

```python
tmp = df.groupby("요일").agg({"지불금액":"mean"}).reset_index()
```

```python
fig,ax = plt.subplots()
ax.plot(tmp["요일"],tmp["지불금액"])
plt.show()
```

#Scatter plot


# Barplot, Barhplot (세로,가로)
- 범주형 변수에 대한 집계를 시각화
- 개수 비교할때 사용

```python
df["요일"].value_counts()
```

```python
fig,ax = plt.subplots()
ax.bar(df["요일"].value_counts().index,df["요일"].value_counts())  # ax.bard 오른쪽
plt.show()
```

- 요일별 저녁타임 식자의 총횟수

```python
tmp = df.groupby("요일").agg({"식사타임":lambda x :(x == "Dinner").sum()}).reset_index()
```

```python
df["식사타임"]
```

```python
tmp = df.groupby("요일").agg({"식사타임": lambda x :(x == "Dinner").sum()}).reset_index()
fig,ax = plt.subplots()
ax.plot(tmp["요일"],tmp["식사타임"])
plt.show()
```

- Pie charts
- 원그래프
- 전체에서 각 부분의 비율을 부채꼴 모양으로 나타낸 그래프

```python
fig,ax = plt.subplots(figsize=(7,5))

ax.pie(
    tmp["식사타임"],
    labels= tmp["요일"],
    startangle =90,
    colors = ["r","g","y","b"],
    explode =[0.1,0,0,0], # 삐져나오게 할 수있음
    autopct = '%.2f%%'
)
plt.show()
```

# Box Plot

```python
np.arange(0,1.1,0.25)
```

```python
q_arr = df["팁금액"].quantile(np.arange(0,1.1,0.25)).to_numpy()
q_arr
```

```python
iqr =q_arr[3] - q_arr[1]

box_min = q_arr[1] - iqr *1.5 # 박스플롯의 최소값
box_max = q_arr[3] + iqr *1.5 # 최대값
iqr,box_min,box_max
```

```python
yticks = np.concatenate([
    q_arr,[iqr,box_min,box_max]
])
yticks
```

```python
fig ,ax = plt.subplots(figsize =(15,10))
ax.boxplot(df["팁금액"])
ax.set_yticks(yticks)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.columns = ["꽃받침길이","꽃받침너비","꽃잎길이","꽃잎너비","꽃종류"]
df.head()
```

# Seaborn
- 내부적으로 Matplotlib 을 호출하여 시각화하는 라이브러리

```python
sns.countplot(df["꽃종류"])
plt.show()
```

```python
sns.countplot(x="꽃종류",data=df).tick_params(axis="x",labelrotation=90)
plt.show()
```

# seaborn 그래프 종류
- Relational plots
    - 두 변수의 관계를 볼때
- Distribution plots
    - 변수의 데이터에 분포 볼때
- Categorical plots
    - 범주형 변수의 집계나 범주형 변수와 수치형 변수간에 관계를 볼때

## Relational plots

- line plot

```python
sns.lineplot(x = "꽃받침길이" , y="꽃잎길이" , data=df)
plt.show()
```

```python
sns.lineplot(x="꽃받침길이",y="꽃잎길이",hue="꽃종류",data=df)
plt.show()
```

```python
sns.scatterplot(x="꽃받침길이",y="꽃잎너비",data=df)
plt.show()
```

- relplot
    - lineplot 과 scatterplot 모두 그릴수 있다.

```python
sns.relplot(x="꽃받침길이",y ="꽃잎너비" ,data = df) # kind = 'line'
plt.show()
```

- col 아규먼트는 열, row 아규먼트는 행

```python
sns.relplot(x="꽃받침길이",y="꽃잎너비",kind="line",data=df,row="꽃종류")
plt.show()
```

- lmplot
    - 추정회귀선과 산점도를 그린다.
    - 추정회귀선이란 x 에 따른 y값의 변화를 잘 설명하는 직선

```python
sns.lmplot(x="꽃받침길이",y="꽃잎너비",data=df)
plt.show()
```

```python
sns.lmplot(x="꽃받침길이",y="꽃잎너비",data=df,hue="꽃종류")
plt.show()
```

- heatmap

```python
sns.heatmap(df.corr(),annot=True,cmap="Blues")
plt.show()
```

- pairplot
    - 대각선 받향으로 히스토그램을 나타낸다
    - 서로 다른 두변수간의 산점도를 그린다.

```python
sns.pairplot(data=df)
plt.show()
```

```python
sns.pairplot(data=df,hue="꽃종류")
plt.show()
```

```python
# 추정회귀선 같이 그려주기
sns.pairplot(data=df,kind="reg")
plt.show()
```

## Distribution plots

- distplot

```python
sns.distplot(df["꽃받침너비"])
plt.show()
```

```python
sns.displot(df["꽃받침너비"]) # kind="kde"
plt.show()
```

- kdeplot

```python
sns.kdeplot(df["꽃받침너비"])
plt.show()
```

```python
sns.kdeplot(x="꽃받침길이",y="꽃받침너비",data=df,shade=True)
plt.show()
```

## Categorical plots

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
cols_rename = {
    "survived" : "생존여부",
    "pclass" : "티켓등급",
    "sex" : "성별",
    "age":"나이",
    "sibsp":"형제자매_배우자수",
    "parch":"부모_자식수",
    "fare": "운임",
    "embarked" : "탑승항구"
}
df = df.rename(columns = cols_rename)
df.head()
```

- countplot

```python
# 성별로 생존여부 카운팅!!
sns.countplot(x="성별",hue="생존여부",data=df)
plt.show()
```

```python
# 티켓등급별로 생존여부 카운팅
sns.countplot(x="티켓등급",hue="생존여부",data=df)
plt.show()
```

- barplot
    - x(범주형)에 대한 y(수치형)의 통계치를 bar 형태로 시각화
    - 신뢰구간도 표시!!

```python
# 성별 생존 비율(수치에 대해서 평균)을 보고싶다면?
sns.barplot(x="성별",y="생존여부",data=df) # 기본값 평균!!
plt.show()
```

```python
# 티켓등급별로 운임료의 평균
sns.barplot(x="티켓등급",y="운임",data=df)
plt.show()
```

```python
# 티켓등급별로 운임료의 표준편차를 보고싶다면?
sns.barplot(x="티켓등급",y="운임",data=df,estimator=np.std)
plt.show()
```

- boxplot

```python
sns.boxplot(x="나이",data=df)
plt.show()
```

```python
# 성별로 나이에 대한 boxplot 보고 싶다면
sns.boxplot(x="성별",y="나이",data=df)
plt.show()
```

```python
# 각 생존여부에 대하여 성별로 운임에대한 boxplot을 보고 싶다면?
sns.boxplot(x="생존여부", hue="성별",y="운임",data=df)
plt.show()
```
